```python
#Official player stats for 2019

players = [{
   'name': 'Julio Jones',
   'catches': 99,
   'targets': 157
   },
   {
   'name': 'Davante Adams',
   'catches': 83,
   'targets': 127
   },
   {
   'name': 'Michael Thomas',
   'catches': 149,
   'targets': 185
   },
]
```

```python

for player in players:
  name = player["name"]
  catches = player["catches"]
  targets = player["targets"]
  catch_rate = catches/targets
  print(name + ' has a catch rate of ' + str(catch_rate))
```

```python

te_df = df.loc[df['Pos'] == 'TE']


te_df = te_df[:30]

te_df = te_df.reset_index(drop=True)


import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# get the frequency on the histogram

freq, bins, patches = plt.hist(te_df['FantasyPoints'], edgecolor='white', label='d')
# x coordinate for labels
bin_centers = np.diff(bins)*0.5 + bins[:-1]

n = 0
for fr, x, patch in zip(freq, bin_centers, patches):
  height = int(freq[n])
  plt.annotate("{}".format(height),
               xy = (x, height),             # top left corner of the histogram bar
               xytext = (0,0.2),             # offsetting label position above its bar
               textcoords = "offset points", # Offset (in points) from the *xy* value
               ha = 'center', va = 'bottom'
               )
  n = n+1

plt.title('Tight Ends')
plt.legend()
plt.show;
```
